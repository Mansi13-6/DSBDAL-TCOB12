```python
import pandas as pd
import numpy as np
from sklearn import preprocessing
from sklearn.datasets import load_iris

iris = load_iris()

df = pd.DataFrame(iris.data, columns=iris.feature_names)

df['Species'] = iris.target

print("First 5 rows:")
print(df.head())

print("\nShape of Dataset:")
print(df.shape)

print("\nDataset Info:")
print(df.info())

print("\nStatistical Summary:")
print(df.describe())

print("\nMissing Values:")
print(df.isnull().sum())

X = df.iloc[:, 0:4].values
print(X)

min_max_scaler = preprocessing.MinMaxScaler()
X_scaled = min_max_scaler.fit_transform(X)
df_scaled = pd.DataFrame(X_scaled, columns=iris.feature_names)
print("\nScaled Data:")
print(df_scaled.head())


df['Species'] = iris.target_names[iris.target]
print("\nUnique Species Names:")
print(df['Species'].unique())


label_encoder = preprocessing.LabelEncoder()
df['Species'] = label_encoder.fit_transform(df['Species'])
print("\nEncoded Species Values:")
print(df['Species'].unique())

print("\nDuplicate Rows Count:")
print(df.duplicated().sum())

print("\nRemoving Duplicates...")
df = df.drop_duplicates()
print("New Shape:", df.shape)


print("\nCorrelation Matrix:")
print(df.corr())


print("\nFilter Sepal Columns:")
print(df.filter(like='sepal'))

print("\nFilter Specific Columns:")
print(df.filter(items=['sepal length (cm)', 'Species']))

print("\nCount of Non-Null Values:")
print(df.count())

print("\nSpecies Value Counts:")
print(df['Species'].value_counts())


print("\nRow at index 5:")
print(df.iloc[5])  
```
