```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

titanic = sns.load_dataset("titanic")
print(titanic.head())

print(titanic.info())

print(titanic.describe())

#Distribution Plot
sns.distplot(titanic["age"], kde=True)
plt.title("Distribution of Age")
plt.show()

#Joint Plot
sns.jointplot(x="age", y="fare", data=titanic, kind="scatter")
plt.show()

#Bar Plot
sns.barplot(x="sex", y="fare", data=titanic)
plt.title("Average Fare by Gender")
plt.show()

#Count Plot
sns.countplot(x="pclass", data=titanic)
plt.title("Passenger Count by Class")
plt.show()

#Box Plot
sns.boxplot(x="pclass", y="fare", data=titanic)
plt.title("Fare Distribution by Class")
plt.show()

#Violin Plot
sns.violinplot(x="pclass", y="fare", data=titanic)
plt.title("Violin Plot of Fare by Class")
plt.show()

#Strip Plot
sns.stripplot(x="pclass", y="age", data=titanic)
plt.title("Strip Plot of Age by Class")
plt.show()

#Swarm Plot
sns.swarmplot(x="pclass", y="age", data=titanic)
plt.title("Swarm Plot of Age by Class")
plt.show()

#Heatmap
plt.figure(figsize=(10,6))
sns.heatmap(titanic.corr(numeric_only=True), annot=True, cmap="coolwarm")
plt.title("Correlation Heatmap")
plt.show()

#Cluster Map
sns.clustermap(titanic.corr(numeric_only=True), annot=True, cmap="coolwarm")
plt.show()

#Histogram
plt.figure(figsize=(8,5))

sns.histplot(titanic["fare"], bins=30, kde=True)

plt.title("Distribution of Ticket Fare")
plt.xlabel("Fare")
plt.ylabel("Number of Passengers")

plt.show()
```
