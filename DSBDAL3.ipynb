```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

df = pd.read_csv("data.csv")
print(df.head())

df.info()

print(df.groupby("Age_Group")["Income"].mean())

print(df.groupby("Age_Group")["Income"].median())

print(df.groupby("Age_Group")["Income"].min())

print(df.groupby("Age_Group")["Income"].max())

print(df.groupby("Age_Group")["Income"].std())

young_income = df[df["Age_Group"] == "Young"]["Income"].tolist()
adult_income = df[df["Age_Group"] == "Adult"]["Income"].tolist()
senior_income = df[df["Age_Group"] == "Senior"]["Income"].tolist()

print("Young Income List:", young_income)
print("Adult Income List:", adult_income)
print("Senior Income List:", senior_income)
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

iris = pd.read_csv("iris.csv")
print(iris.head())

print(iris["Species"].unique())

print(iris.groupby("Species").mean())

print(iris.groupby("Species").std())

print(iris.groupby("Species").describe())

setosa = iris[iris["Species"] == "Iris-setosa"]
print(setosa.describe())

versicolor = iris[iris["Species"] == "Iris-versicolor"]
print(versicolor.describe())

virginica = iris[iris["Species"] == "Iris-virginica"]
print(virginica.describe())
```
